# Stage 08 - Exploratory Data Analysis

Fill in the marked TODOs. This notebook generates synthetic data so you can focus on the EDA flow. Replace with your dataset when ready.

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from scipy.stats import skew, kurtosis
sns.set(context='talk', style='whitegrid')
np.random.seed(8)
pd.set_option('display.max_columns', 100)

# === Synthetic data generator (adapt or replace with your own data) ===
n = 160
df = pd.DataFrame({
    'date': pd.date_range('2021-02-01', periods=n, freq='D'),
    'region': np.random.choice(['North','South','East','West'], size=n),
    'age': np.random.normal(40, 8, size=n).clip(22, 70).round(1),
    'income': np.random.lognormal(mean=10.6, sigma=0.3, size=n).round(2),
    'transactions': np.random.poisson(lam=3, size=n),
})
base = df['income'] * 0.0015 + df['transactions']*18 + np.random.normal(0, 40, size=n)
df['spend'] = np.maximum(0, base).round(2)

# inject a bit of missingness and outliers
df.loc[np.random.choice(df.index, 5, replace=False), 'income'] = np.nan
df.loc[np.random.choice(df.index, 3, replace=False), 'spend'] = np.nan
df.loc[np.random.choice(df.index, 2, replace=False), 'transactions'] = df['transactions'].max()+12
df.head()
```

## Part 1 - First look

```python
df.info(), df.isna().sum()
```

## Part 2 - Numeric profile

```python
desc = df[['age','income','transactions','spend']].describe().T
desc['skew'] = [skew(df[c].dropna()) for c in desc.index]
desc['kurtosis'] = [kurtosis(df[c].dropna()) for c in desc.index]
desc
```

## Part 3 - Distributions

```python
# Histogram + KDE for income
sns.histplot(df['income'], kde=True)
plt.title('Income Distribution')
plt.show()

# Histogram + KDE for age
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()

# Histogram + KDE for spend
sns.histplot(df['spend'], kde=True)
plt.title('Spend Distribution')
plt.show()

# TODO: boxplot for outliers
sns.boxplot(x=df['transactions'])
plt.title('Transactions (Outliers)')
plt.show()
```

## Part 4 - Relationships

```python
sns.scatterplot(data=df, x='income', y='spend', hue='region')
plt.title('Income vs Spend')
plt.show()

sns.scatterplot(data=df, x='age', y='spend')
plt.title('Age vs Spend')
plt.show()
```

## Part 5 - Correlation matrix

```python
corr = df[['age','income','transactions','spend']].corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt='.2f', cmap='vlag', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
corr
```

## Part 6 Insights & Assumptions

### Insights

- It's easy to notice a rightward skewness in the income distribution, which means that there exist people with extremely high income and that the mean is less than median. Similarly, a slight skewness to the right can be noticed in the spend distribution. However, no significant skewness can be identified in the age distribution. 
- Outliers could be identified in the transactions, which might have a great impact on the overall distribution. 
- It's easy to identify a positive correlation between income and spend from both the scatterplot and the correlation matrix. 

### Assumptions & Risks

- Linearity is assumed of transactions and income's impacts on spend. If non-linearities exist (e.g. from incomes skewed distribution), the models would be biased.
- The outliers in transactions can distort impact coefficients. 
- Variance in spend may increase with transactions or income: should check residuals and consider inference robust to heteroskedasticity. 

### Next Steps

- Handle missing values: replace with mean/median or delete the whole row. 
- Treat outliers: cap them or use winsorization. 
- Normalization: eliminate any skewness of raw column distributions. 
- Validation: use train-test split or testing using regression models.
