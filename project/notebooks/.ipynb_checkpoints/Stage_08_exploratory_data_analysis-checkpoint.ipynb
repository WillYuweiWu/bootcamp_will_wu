# Stage 08 - Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

df = pd.read_csv('../data/processed/VIX_S&P500_clean.csv')
df.info()
```

```python
df.head()
```

```python
# Missing values

df.isna().sum()
```

```python
# Numeric profile

num_cols = df.select_dtypes(include = [np.number]).columns
desc = df[num_cols].describe().T
desc['skew'] = [skew(df[c].dropna()) for c in desc.index]
desc['kurtosis'] = [kurtosis(df[c].dropna()) for c in desc.index]
display(desc)
```

```python
# Boxplots and histograms of vix_close and sp500_close

fig, axes = plt.subplots(2,2, figsize = (20,10))

sns.boxplot(x = df['vix_close'], ax = axes[0,0])
axes[1,0].set_title('VIX Close Boxplot')

sns.boxplot(x = df['sp500_close'], ax = axes[0,1])
axes[1,1].set_title('S&P 500 Close Boxplot')

sns.histplot(df['vix_close'], kde = True, ax = axes[1,0])
axes[0,0].set_title('VIX Close Histogram')

sns.histplot(df['sp500_close'], kde = True, ax = axes[1,1]) 
axes[0,1].set_title('S&P 500 Close Histogram')

plt.show()
```

```python
# Scatterplot

plt.figure(figsize = (15,5))
sns.scatterplot(data = df, x = 'vix_close', y = 'sp500_close', alpha = 0.5)
plt.title('VIX v.s. S&P 500')
plt.show()
```

```python
# Correlation heatmap

corr = df[num_cols].corr(numeric_only = True)
sns.heatmap(corr, annot = True, fmt = '.2f', cmap = 'vlag', vmin = -1, vmax = 1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Additional trends over time

plt.figure(figsize = (20,5))
plt.plot(df['date'], df['vix_close'], label = 'VIX Close')
plt.title('VIX over Time')
plt.xlabel('Date')
plt.ylabel('VIX')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (20,5))
plt.plot(df['date'], df['sp500_close'], label = 'S&P 500 Close')
plt.title('S&P 500 over Time')
plt.xlabel('Date')
plt.ylabel('S&P 500')
plt.legend()
plt.show()
```
